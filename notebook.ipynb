```python
board = cv2.imread('./templates/board.png')

light_square = np.array([220, 235, 238])
dark_square = np.array([100, 155, 115])
tolerance = 25

light_mask = cv2.inRange(board, 
                         light_square + tolerance,
                         light_square - tolerance)
dark_mask = cv2.inRange(board,
                        dark_square + tolerance,
                        dark_square - tolerance)

board_mask = cv2.bitwise_or(light_mask, dark_mask)

kernel = np.ones((5, 5), np.uint8)
board_mask = cv2.morphologyEx(board_mask, cv2.MORPH_CLOSE, kernel)
```
